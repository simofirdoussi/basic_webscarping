```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r = requests.get("https://www.avito.ma/fr/maroc/voitures-%C3%A0_vendre?cb=13&mo=logan&fl=1", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c = r.content
```

```python
soup = BeautifulSoup(c,"html.parser")
```

```python
titles = soup.find_all("h2",{"class":"fs14"}) #extracting the titles
```

```python
cities = soup.find_all("span",{"class":"item-info-extra fs14"}) #Extracting the cities names
```

```python
price_values = soup.find_all("span",{"class":"price_value"}) #Extracting the price values of each title
```

```python
l = []
for i in range (len(titles)):
    d = {}
    d["Type"] = (titles[i].text)
    d["City"] = (cities[i].text.replace("\n","").replace(' -Annonce de professionnel', ''))
    d["Price"] = (price_values[i].text)
    if d["Price"] == "\xa0":# Condition for the case where the price isnt specified
        d["Price"] = "NONE"
    
    l.append(d)
```

```python
l 
```

```python
df = pd.DataFrame(l) #creating a dataframe
df
```

```python
df.to_csv("Output.csv") #storing the poutput in a csv file 
```
